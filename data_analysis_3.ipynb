To separate in subsets i need to use PCA to reduce the dimensionality of the data set

```python
## I need to define exactly the data that i going to use for PCA
X= data_breast.iloc[:,0:47]
X
```

```python
##It is needed to import the library
import sklearn.decomposition
```

```python
### Initiate de PCA and apply it to our data
pca = sklearn.decomposition.PCA()
pca.fit(X)
pca.n_components_
sns.barplot(x=np.arange(1,48), y=pca.explained_variance_ratio_)
plt.title("PCA Explained variance - Breast Dataset")
plt.xlabel("Principal Components")
plt.ylabel("Percentage of explained variance")
sns.despine()
```

```python
### Apply the dimensionality reduction to our data
reduce_breast = pca.transform(X)

### Make a dataframe and add the label column
reduce_breast = pd.DataFrame(reduce_breast)
#reduce_breast[""]
```

```python
##plot the results ofr the reduction
sns.scatterplot(data=reduce_breast, x=0, y=1)
plt.title("PCA")
plt.xlabel("PCA1")
plt.ylabel("PCA2")
sns.despine()
```

Once with the dimension reduced, i can try to find if the dataset has similar objects. For that, i need a K (quantity of clusters), but i do not  have a "type" into my dataset. So, i am going to use the "elbow" method to decide the value of K. 

```python
## to cluster is needed to import
import sklearn.cluster
```

```python
##set the datase
X=reduce_breast.iloc[:, 0:47]

### Run kmeans for various k values
### initiate empty list
inertia_list = []

### Create for loop for K from 1 to 10
for k in range(1,11):
    kmeans = sklearn.cluster.KMeans (k)
    kmeans = kmeans.fit(X)
    
    inertia_list += [kmeans.inertia_]
```

```python
#checking
inertia_list
```

```python
### Elbow plot
sns.scatterplot(x = np.arange(1,11), y=inertia_list)
sns.lineplot(x =np.arange(1,11), y=inertia_list)
plt.title("Elbow graph")
plt.xlabel("k")
plt.ylabel("inertia")
sns.despine()
```

At the first moment i decide to take k=7 and k=4 to compare the behavior of the clusters. What i suppose it is that maybe, k=7 i going to help me visualize overlap regions better than k=4. 

```python
## changing the number of clusters
k=4
kmeans = sklearn.cluster.KMeans (k)
kmeans = kmeans.fit(X)
reduce_breast["cluster"]=kmeans.labels_
reduce_breast

#see the clusters
sns.scatterplot(data=reduce_breast, x=0, y=1,hue ="cluster")
plt.title("Cluster - expression dataset k=4")
plt.xlabel("PCA2 ")
plt.ylabel("PCA1")
sns.despine()
```

```python
## changing the number of clusters
k=7
kmeans = sklearn.cluster.KMeans (k)
kmeans = kmeans.fit(X)
reduce_breast["cluster"]=kmeans.labels_
reduce_breast

#see the clusters
sns.scatterplot(data=reduce_breast, x=0, y=1,hue ="cluster")
plt.title("Cluster - expression dataset k=7")
plt.xlabel("PCA2 ")
plt.ylabel("PCA1")
sns.despine()
```

```python
## changing the number of clusters
k=7
kmeans = sklearn.cluster.KMeans (k)
kmeans = kmeans.fit(X)
reduce_breast["cluster"]=kmeans.labels_
reduce_breast

#see the clusters
sns.relplot(data=reduce_breast, x=0, y=1,col ="cluster")
plt.title("Cluster - expression dataset k=7")
plt.xlabel("PCA2 ")
plt.ylabel("PCA1")
sns.despine()
```

In k=7 it is possible to observe overlap regions in the range (-5/2.5) and range (5/7.5). An interesting pattern arose. Even with differents k the leftmost cluster remains almost intact, not blending in with the others (in k=4 is light pink, in k=7 is pink).
